###Python Comments

Comments in Python are lines in the code that are not executed by the computer. They are written to help explain what the code does, making it easier for others (or your future self) to understand. Comments are a crucial part of writing readable, maintainable code.

Python supports single-line comments using the # symbol
Multi-line comments using triple quotes (''' or """).

Comments can be used to explain Python code.

Comments can be used to make the code more readable.

Comments can be used to prevent execution when testing code.

✅ Real-life example: If it's a holiday, check whether it's raining or sunny.

```python
is_holiday = input("Is it a holiday? (yes/no): ").lower()

if is_holiday == "yes":
    weather = input("What's the weather like? (rainy/sunny): ").lower()

    if weather == "rainy":
        print("Stay home and watch a movie 🎬")
    elif weather == "sunny":
        print("Go out and enjoy the sun ☀️")
    else:
        print("Enjoy your day however you like! 😎")
else:
    print("Get ready for work or school! 🏫💼")
```

# 1. while loop

📘 What is a while loop?
A while loop repeats something again and again as long as a condition is true.

We use it when we don't know how many times we need to repeat.



---


example:

Imagine your water bottle is half full. You will keep drinking while the bottle is not empty.

So, the condition is:

“While bottle is not empty → keep drinking”


image.png

**Example 1: Print numbers 1 to 5**

```python
i = 1
while i <= 5:
    print("Counting:", i)
    i = i + 1
```

**Example 2: Keep asking until user enter correct password**

```python
password = 12345
attempt = ""

while attempt != password:
    attempt = int(input("Enter password: "))
    if attempt != password:
        print("Wrong Password! Try Again")

print("Access Granted")
```

# **2. for loop**

What is a for loop?


A for loop lets us repeat something again and again automatically.

We use it when we know how many times we want to repeat.



---


example:

Imagine you have a box with 5 apples.
You want to check each apple one by one.

You don’t say:

"Pick first apple… Pick second apple… Pick third apple…"

Instead, you say:

"For every apple in the box, pick it and check it."


That's exactly what a for loop does!

for-loop-python.jpg



**Example 1: Print numbers 1 to 5**

```python
for i in range(1,10):  # Starts from 1 and goes till 5
    print("Number is:", i)
```

**Example 2: Print items in a list**

```python
fruits = ["apple", "banana", "mango" , "cheery" , "pineapple", "watermellon"]
# for fruit in fruits:
#     print("I love", fruit)

for i in range(0,len(fruits)): # start, end, increment
  print(fruits[i])
```

**Example 3: Loop with index (show item and its number)**

```python
colors = ["red", "green", "blue"]
for i, color in enumerate(colors):
    print("Color number", i+1, "is", color)
```
