# **1. for loop**

What is a for loop?


A for loop lets us repeat something again and again automatically.

We use it when we know how many times we want to repeat.



---


example:

Imagine you have a box with 5 apples.
You want to check each apple one by one.

You don’t say:

"Pick first apple… Pick second apple… Pick third apple…"

Instead, you say:

"For every apple in the box, pick it and check it."


That's exactly what a for loop does!

for-loop-python.jpg



# Example 1: Print numbers 1 to 5

```python
for i in range(1,10,2):  # Starts from 1 and goes till 5
    print("Number is:", i)
```

Explanation:

range(1, 6) means: 1, 2, 3, 4, 5 (it stops before 6)

It prints each number in the loop.



# Example 2: Print items in a list

```python
fruits = ["apple", "banana", "mango" , "cheery" , "pineapple", "watermellon"]
for fruit in fruits:
    print("I love", fruit)

#for i in range(0,len(fruits),2):
  #print(fruits[i])
```

Explanation:

We created a list of fruits.

The loop goes to each item one by one and prints it.

# Example 3: Loop with index (show item and its number)

```python
colors = ["red", "green", "blue"]
for i, color in enumerate(colors):
    print("Color number", i, "is", color)
```

Explanation:

enumerate helps us get both the index (i) and the item (color).

# 2. while loop

📘 What is a while loop?
A while loop repeats something again and again as long as a condition is true.

We use it when we don't know how many times we need to repeat.



---


example:

Imagine your water bottle is half full. You will keep drinking while the bottle is not empty.

So, the condition is:

“While bottle is not empty → keep drinking”


image.png

# Example 1: Print numbers 1 to 5

```python
i = 1
while i <= 5:
    print("Counting:", i)
    i = i + 1
```

Explanation:

We start with i = 1.

It prints and increases i until it becomes 6.

# Example 2: Keep asking until user types exit

```python
while True:
    word = input("Type 'exit' to stop: ")
    if word == "exit":
        break
    print("You typed:", word)
```

Explanation:

It keeps asking for input forever until user types "exit".

break stops the loop.

# 3. break and continue
📘 What is break?
break is used to stop a loop immediately.



# 📘 What is continue?
continue is used to skip one step in the loop but keep going.

# examples:
**break:** You are searching for your keys in drawers. As soon as you find them, you stop checking further drawers.

**continue:** You are cleaning your room but skip the drawer with your personal items.

# break Example

```python
for num in range(1, 10):
    if num == 5:
        print("Found 5, breaking the loop.")
        break
    print("Current number:", num)
```

# continue Example

```python
for num in range(1, 6):
    if num == 3:
        # print("Skipping 3")
        continue
    print("Number is:", num)
```

# Common List Methods
Here is a table of common list methods:

| Method          | Description                            | Example                   |
| --------------- | -------------------------------------- | ------------------------- |
| `.append(x)`    | Adds an item at the **end**            | `list.append("apple")`    |
| `.insert(i, x)` | Inserts an item at **index i**         | `list.insert(1, "grape")` |
| `.remove(x)`    | Removes **first occurrence** of item x | `list.remove("apple")`    |
| `.pop()`        | Removes the **last item**              | `list.pop()`              |
| `.pop(i)`       | Removes item at index i                | `list.pop(2)`             |
| `.clear()`      | Removes **all items**                  | `list.clear()`            |
| `.index(x)`     | Returns the index of item x            | `list.index("apple")`     |
| `.count(x)`     | Counts how many times x appears        | `list.count("apple")`     |
| `.sort()`       | Sorts the list in **ascending** order  | `list.sort()`             |
| `.reverse()`    | Reverses the list order                | `list.reverse()`          |
| `len(list)`     | Returns number of items                | `len(list)`               |
| `in`            | Checks if item exists                  | `"apple" in list`         |


# 📌 append() – Add at the end

```python
fruits = ["apple", "banana"]
print(fruits)
fruits.append("mango")
print(fruits)  # ['apple', 'banana', 'mango']
```

# 📌 insert(index, item) – Add at specific position

```python
fruits = ["apple", "banana"]
print(fruits)
fruits.insert(1, "orange")
print(fruits)  # ['apple', 'orange', 'banana']
```

# 📌 remove(item) – Remove by value

```python
fruits = ["apple", "banana"]
print(fruits)
fruits.remove("banana")
print(fruits)  # ['apple']
```

# 📌 pop() – Remove last item

```python
fruits = ["apple", "banana", "mango"]
print(fruits)
last_fruit = fruits.pop()
print(last_fruit)  # mango
print(fruits)      # ['apple', 'orange']
```

# 📌 pop(index) – Remove specific index

```python
fruits = ["apple", "banana", "mango"]
print(fruits)
fruits.pop(0)      # removes 'apple'
print(fruits)      # ["banana", "mango"]
```

# 📌 clear() – Remove all items

```python
fruits = ["apple", "banana", "mango"]
print(fruits)
fruits.clear()
print(fruits)  # []
```

# 📌 index() – Find index of item

```python
colors = ["red", "green", "blue"]
print(colors.index("green"))  # 1
```

# 📌 count() – Count occurrences

```python
nums = [1, 2, 3, 2, 2, 4]
print(nums.count(2))  # 3
```

# 📌 sort() – Sort in order

```python
numbers = [5, 3, 9, 1]
numbers.sort()
print(numbers)  # [1, 3, 5, 9]
```

# 📌 reverse() – Reverse the list

```python
numbers = [5, 3, 9, 1]
numbers.reverse()
print(numbers)  # [9, 5, 3, 1]
```

# 📌 len() – Count items

```python
numbers = [5, 3, 9, 1]
print(len(numbers))  # 4
```

# 📌 in – Check if item is in list

```python
fruits = ["apple", "banana"]
if "apple" in fruits:
    print("Yes, apple is in the list!")
else:
  print("No, apple is not in the list")
```

# ✅ Looping Through Lists

**🔁 Using for loop**

```python
colors = ["red", "green", "blue"]
for color in colors:
    print("Color:", color)
```

**🔁 Using while loop**

```python
colors = ["red", "green", "blue"]
i = 0
while i < len(colors):
    print(colors[i])
    i += 1
```

# ✅ Advanced: List Comprehension

**⚡ Fast way to create lists**

```python
squares = [x**2 for x in range(1, 10 )]
print(squares)  # [1, 4, 9, 16, 25, 36, 49, 64, 81]
```

# 🛒 Python Mart – Console-Based Shopping Cart Project

Description:

"Python Mart" is a beginner-friendly console project that simulates a real-life shopping experience using Python lists and loops. Users can view products, add or remove items from a shopping cart, and checkout. This project helps to understand core programming concepts like loops, conditional statements, indexing, list manipulation, and user input handling. It’s a great hands-on introduction to how Python can be used to build simple real-world applications.

```python
# Product Data
products = ["Laptop", "Mouse", "Keyboard", "Charger", "Mobile"]
prices = [70000, 1500, 3000, 2000, 30000]
cart = []

# Welcome Message
print("🛍️ Welcome to Python Mart!\n")

# Main Menu Loop
while True:
    print("\nPlease select an option:")
    print("1. View Products")
    print("2. Add to Cart")
    print("3. View Cart")
    print("4. Remove from Cart")
    print("5. Checkout")
    print("6. Exit")

    choice = input("Enter your choice (1–6): ")

    # Option 1: View Products (with Submenu)
    if choice == "1":
        while True:
            print("\n🛒 Available Products:")
            for i in range(len(products)):
                print(f"{i+1}. {products[i]} - Rs.{prices[i]}")

            print("\nOptions:")
            print("a. Add a product to cart")
            print("b. Back to main menu")

            sub_choice = input("Choose an option (a/b): ").lower()

            if sub_choice == "a":
                try:
                    item_no = int(input("Enter product number to add to cart: ")) - 1
                    if 0 <= item_no < len(products):
                        cart.append((products[item_no], prices[item_no]))
                        print(f"✅ {products[item_no]} added to cart.")

                        # Show updated cart
                        print("\n🛒 Your Updated Cart:")
                        total = 0
                        for i in range(len(cart)):
                            print(f"{i+1}. {cart[i][0]} - Rs.{cart[i][1]}")
                            total += cart[i][1]
                        print(f"Total: Rs.{total}")
                    else:
                        print("❌ Invalid product number.")
                except ValueError:
                    print("❌ Please enter a valid number.")
            elif sub_choice == "b":
                break
            else:
                print("❗ Invalid option. Please enter 'a' or 'b'.")

    # Option 2: Add to Cart (Direct)
    elif choice == "2":
        try:
            for i in range(len(products)):
                print(f"{i+1}. {products[i]} - Rs.{prices[i]}")
            item_no = int(input("Enter product number to add to cart: ")) - 1
            if 0 <= item_no < len(products):
                cart.append((products[item_no], prices[item_no]))
                print(f"✅ {products[item_no]} added to cart.")

                # Show updated cart
                print("\n🛒 Your Updated Cart:")
                total = 0
                for i in range(len(cart)):
                    print(f"{i+1}. {cart[i][0]} - Rs.{cart[i][1]}")
                    total += cart[i][1]
                print(f"Total: Rs.{total}")
            else:
                print("❌ Invalid product number.")
        except ValueError:
            print("❌ Please enter a valid number.")

    # Option 3: View Cart
    elif choice == "3":
        if not cart:
            print("🛒 Your cart is empty.")
        else:
            print("\n🧾 Your Cart:")
            total = 0
            for i in range(len(cart)):
                print(f"{i+1}. {cart[i][0]} - Rs.{cart[i][1]}")
                total += cart[i][1]
            print(f"Total: Rs.{total}")

    # Option 4: Remove from Cart
    elif choice == "4":
        if not cart:
            print("❌ Cart is empty.")
            continue
        for i in range(len(cart)):
            print(f"{i+1}. {cart[i][0]} - Rs.{cart[i][1]}")
        try:
            remove_index = int(input("Enter item number to remove: ")) - 1
            if 0 <= remove_index < len(cart):
                removed_item = cart.pop(remove_index)
                print(f"❌ {removed_item[0]} removed from cart.")
            else:
                print("❌ Invalid item number.")
        except ValueError:
            print("❌ Please enter a valid number.")

    # Option 5: Checkout
    elif choice == "5":
        if not cart:
            print("🛒 Your cart is empty.")
        else:
            print("\n🧾 Final Bill:")
            total = 0
            for item in cart:
                print(f"{item[0]} - Rs.{item[1]}")
                total += item[1]
            print(f"Total Amount: Rs.{total}")
            print("✅ Thank you for shopping with us!")
        break

    # Option 6: Exit
    elif choice == "6":
        print("👋 Exiting without checkout. Bye!")
        break

    # Invalid Input
    else:
        print("❗ Invalid input. Please choose from 1 to 6.")
```
